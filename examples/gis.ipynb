```r
library("jupytergis")
```

```r
doc <- GISDocument$new("france_hiking.jGIS")
doc
```

```r
google_layer <- doc$add_raster_layer(
    url = "https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}",
    name = paste("Google Satellite", format(Sys.time(), "%H:%M:%S")),
    attribution = "Google",
    opacity = 0.6
)
```

```r
# retrieve opacity
doc$with_read(function(t) {
    doc$layers$get(t, google_layer)$parameters$opacity
})
```

```r
osm_layer <- doc$add_raster_layer(
    url = "https://tile.openstreetmap.org/{z}/{x}/{y}.png",
    name = paste("OSM from R", format(Sys.time(), "%H:%M:%S")),
)
```
